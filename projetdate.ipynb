# Introduction à pandas


```python
import pandas as pd # pd est un alias couramment utilisé pour pandas

data = {
    'Annee': [2024, 2025, 2026, 2027],
    'Diplomes': ['bac', 'B1', 'B2', 'B3']

}

df = pd.DataFrame(data) # Création d'un DataFrame à partir du dictionnaire de données
print(df)
```

```python
df["Annee"].max()  # Trouver l'année maximale dans la colonne 'Annee'
```

```python
df.describe()  # Obtenir des statistiques descriptives du DataFrame
```

```python
mouennes = {
    'Annee': [2024, 2025, 2026, 2027],
    'Diplomes': ['bac', 'B1', 'B2', 'B3'],
    'Moyennes': [12.5, 13.0, 14.2, 15.1]
}

df_moyennes = pd.DataFrame(mouennes) # Création d'un DataFrame à partir du dictionnaire de données
df_moyennes
```

```python
import matplotlib.pyplot as plt
plt.plot(df_moyennes['Annee'], df_moyennes['Moyennes'], marker='o') # Tracer les moyennes par année
plt.title('Moyennes des Diplômes par Année')
plt.xlabel('Année')
plt.ylabel('Moyenne')
plt.grid()

```
